```python
import geopandas as gpd
from matplotlib import pyplot as plt
from shapely import Point, Polygon, unary_union


# Read in files and make column names descriptive
da_home = gpd.read_file('./Dissemination area - Homeownership rate (%)/Dissemination area - Homeownership rate (%).shp')
da_home.rename(columns={'Value': 'Percentage of Home Ownership 2021'}, inplace=True)
```

```python
da_inc = gpd.read_file('./Dissemination area - Median total income of households in 2020 ($)/Dissemination area - Median total income of households in 2020 ($).shp')
da_inc.rename(columns={'Value': 'Median Household Income 2020'}, inplace=True)
```

```python
da_min = gpd.read_file('./Dissemination area - Percentage of visible minority/Dissemination area - Percentage of visible minority, n.i.e..shp')
da_min.plot(column = 'Value', scheme='quantiles')
da_min.rename(columns={'Value': 'Percentage of Visible Minorities 2021'}, inplace=True)
```

```python
#da_min['Percentage of Visible Minorities 2021'] = da_min['Value']


da_inc.plot(column='Median Household Income 2020', scheme='BoxPlot')
da_home.plot(column='Percentage of Home Ownership 2021', scheme='BoxPlot')
print(da_min.columns)
#da_min.plot(column= 'Percentage of Visible Minorities 2021', scheme='BoxPlot')

da_gdf = da_min.merge(da_inc, on = 'DGUID')
da_gdf= da_gdf.merge(da_home, on = 'DGUID')
# da_gdf = da_gdf[['DGUID', 'geometry', 'test',
#                  'Median Household Income 2020',
#                  'Percentage of Home Ownership 2021']]
print(da_gdf.columns)
```
